```python
import numpy as np
import matplotlib.pyplot as plt

# Define the original PDF (Standard Normal Distribution)
def pdf_X(x):
    return (1 / np.sqrt(2 * np.pi)) * np.exp(-x**2 / 2)

# Transformation: Y = 2X + 3
def g(x):
    return 2 * x + 3

# Inverse Transformation: X = (Y - 3) / 2
def g_inverse(y):
    return (y - 3) / 2

# Jacobian of the Inverse Transformation
def jacobian_inverse(y):
    return 1 / 2  # |d/dy of g_inverse(y)|

# Transformed PDF of Y
def pdf_Y(y):
    x = g_inverse(y)
    jacobian = jacobian_inverse(y)
    return pdf_X(x) * jacobian

# Generate values for X and Y
x_values = np.linspace(-4, 4, 500)
y_values = np.linspace(-4, 10, 500)

# Compute the PDFs
pdf_X_values = pdf_X(x_values)
pdf_Y_values = pdf_Y(y_values)

# Plot the PDFs
plt.figure(figsize=(10, 6))
plt.plot(x_values, pdf_X_values, label=r"$f_X(x)$: Standard Normal")
plt.plot(y_values, pdf_Y_values, label=r"$f_Y(y)$: Transformed PDF ($Y = 2X + 3$)", linestyle="--")
plt.axvline(x=3, color='gray', linestyle=':', label="Mean Shift to $3$")
plt.title("Transformation of a Standard Normal Distribution")
plt.xlabel("Value")
plt.ylabel("Probability Density")
plt.legend()
plt.grid()
plt.show()
```
